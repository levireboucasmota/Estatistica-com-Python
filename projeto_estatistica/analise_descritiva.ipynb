### Para avaliarmos o comportamento da variável RENDA vamos construir uma tabela de frequências considerando as seguintes classes em salários mínimos (SM)
#### <font color='blue'>Descreva os pontos mais relevantes que você observa na tabela e no gráfico.</font>

Classes de renda:

<b>A</b> ► Acima de 25 SM

<b>B</b> ► De 15 a 25 SM

<b>C</b> ► De 5 a 15 SM

<b>D</b> ► De 2 a 5 SM

<b>E</b> ► Até 2 SM

Para construir as classes de renda considere que o salário mínimo na época da pesquisa era de <b>R$ 788,00</b>.

#### Siga os passos abaixo:

### 1º Definir os intevalos das classes em reais (R$)

```python
classes = [0, 1576, 3940, 11820, 19700, 200000]
```

### 2º Definir os labels das classes

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

### 3º Construir a coluna de frequências

```python
frequencia = pd.Series(
    pd.cut(x=dados.Renda,
           bins=classes,
           labels=labels,
           include_lowest=True).value_counts()
)
frequencia
```

### 4º Construir a coluna de percentuais

```python
percentual = pd.Series(
    pd.cut(x=dados.Renda,
           bins=classes,
           labels=labels,
           include_lowest=True,).value_counts(normalize=True)*100
)
percentual
```

### 5º Juntar as colunas de frequência e percentuais e ordenar as linhas de acordo com os labels das classes

```python
pd.DataFrame({'Frequência': frequencia, 'Porcentagem (%)': percentual}).sort_index(ascending=False)
```

### Construa um gráfico de barras para visualizar as informações da tabela de frequências acima

```python
plot = sns.barplot(x=percentual.index, y=percentual)
plot.set(ylabel='Percentual')
plot
```

> ### Conclusões

Podemos ver com o gráfico de distribuição de renda com uma predominância de pessoas na classe E (Ganham até 2 salários mínimos), representando cerca de 65% do total. Uma outra parte menor está dividida nas classes D (cerca de 25%) e C (cerca de 10%). As classes B e A (Ganhando mais de 15 salários mínimos) representam uma quantidade muito baixa, próxima a 1% ou menos, indicando uma alta concentração de renda na base.

### Obtenha a média, mediana e valor máximo da variável RENDA segundo SEXO e COR
#### <font color='blue'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mean', 'median', 'max'}</i></font>

```python
medidas_tendência_central  = {'mean', 'median', 'max'}
frequencia = pd.crosstab(dados.Sexo, dados.Cor, aggfunc = medidas_tendência_central, values = dados.Renda)
frequencia.rename(index=sexo, inplace=True)
frequencia.rename(columns=cor, inplace=True)
frequencia
```

> ### Conclusões

A renda máxima mais alta da amostra é de 200.000 para homens na categoria Branca. Além disso, os homens tendem a ter uma renda máxima maior que a das mulheres.

Podemos analisar também que a maior média de salário é a dos homens, sendo superior que o das mulheres em todas as cores. Além disso, a categoria Amarela tem um salário médio superior ao das outras.

Temos também a mediana dos salários representando uma renda central, com os homens possuindo uma mediana maior, sendo a maior na categoria Amarela.

*Resumo Geral:*

1. Homens: Em geral, os homens têm uma mediana de renda consistentemente mais alta do que as mulheres em todas as categorias de cor, indicando que a mulher típica ganha menos do que o homem típico.

2. Cor/Raça: As categorias Branca e Amarela apresentam as maiores medianas e médias de renda para ambos os sexos, enquanto Indígena, Preta e Parda apresentam os valores mais baixos.

### Obtenha as medidas de dispersão da variável RENDA segundo SEXO e COR
#### <font color='blue'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mad', 'var', 'std'}</i></font>

```python
def mad_custom(x):
    return np.mean(np.abs(x - np.mean(x)))

medidas_dispersao = {mad_custom, 'var', 'std'}
frequencia = pd.crosstab(dados.Sexo, dados.Cor, aggfunc = medidas_dispersao, values = dados.Renda)
frequencia.rename(index=sexo, inplace=True)
frequencia.rename(columns=cor, inplace=True)
frequencia
```

> ### Conclusões

As mulheres indígenas (std = 11.957,5): O desvio padrão é o maior de toda a tabela. Isso é um forte indício de que, embora a maioria das mulheres Indígenas possa ter uma renda baixa (mediana de 788,0 na análise anterior), a presença de um ou poucos valores de renda extremamente altos (como o máximo de 120.000) distorce a dispersão do grupo como um todo.

Já os homens amarelos (std = 5.740,8): Apresentam a maior dispersão entre os homens.

> ### Conclusões

O método <i>agg()</i> pode receber um dicionário especificando qual coluna do DataFrame deve ser utilizada e qual lista de funções estatísticas queremos obter, por exemplo: <i>dados.groupby(['UF']).agg({'Renda': ['mean', 'median', 'max', 'std']})</i></font>
